```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

gdp = pd.read_csv("gdp_per_capita.csv",thousands=',',delimiter='\t',
                  encoding='latin1',na_values="n/a")

oecd = pd.read_csv("oecd_bli_2015.csv")
```

```
gdp.info()
```

```
gdp.head()
```

```
gdp_min = gdp[["Country","2015"]]
gdp_min.head()
```

```
oecd.head()
```

```
oecd_min = oecd[["Country","Value"]]
oecd_min.head()
```

```
life_sat = pd.merge(gdp_min, oecd_min, on="Country")
life_sat.info()
```

```
%matplotlib inline
life_sat.plot(kind="scatter",x="2015",y="Value")
plt.xlabel("GDP")
plt.ylabel("Life Satisfaction")
plt.title("GDP Vs. Life Satisfaction")
plt.show()
```

# Linear Model Assumption  
We will take an assumption that Life Satisfaction is a linear function of GDP per capita.

$$life\_satisfaction=\theta_0+\theta_1\times GDP$$

Here, $\theta_0$ is bias and $\theta_1$ is the weight of GDP in linear model.

```
import seaborn as sns

sns.regplot(x="2015", y="Value",data=life_sat)
plt.xlabel("GDP")
plt.ylabel("Life Satisfaction")
plt.title("GDP Vs. Life Satisfaction")
plt.show()
```

```
from sklearn.linear_model import LinearRegression

clf = LinearRegression()

X = np.c_[life_sat["2015"]]
y = np.c_[life_sat["Value"]]

clf = clf.fit(X,y)
```

```
X_new=[[22587]]# Cyprus' GDP per capita
print(clf.predict(X_new))# outputs [[ 5.96242338]]
```
